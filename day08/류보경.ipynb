### 2. P269 유명 맛집 대기줄 구현하기 문제를 while True:문을 사용하여 동적으로 입력을 받는 프로그램으로 변경작성하시오.


```python
SIZE = 5
queue = [None] * SIZE
front = rear = -1  


# 큐가 다 찼는지 확인
def isQueueFull(): 
    global SIZE, queue, front, rear
   
    if ((rear+1) % SIZE == front): 
        return True
    else:
        return False

# 큐가 비어있는지 확인
def isQueueEmpty():
    global SIZE, queue, front, rear
    if (front == rear): 
        return True
    else:
        return False
    
# 큐에 값 입력
def enQueue(data):
    global SIZE, queue, front, rear
    if isQueueFull():
        print('Queue is full')
        return
    rear += 1 # 인덱스를 한칸뒤로 해준다음에
    queue[rear] = data # 값 입력

def deQueue():
    global SIZE, queue, front, rear
    if isQueueEmpty():
        print('Queue is empty')
        return None
    front += 1
    data = queue[front] 
    queue[front] = None  # 데이터 삭제

    # 값을 앞으로 땡겨오기 위한 반복문 // None이 뒤에서 부터
    for i in range(front + 1, rear +1):
        queue[i-1] = queue[i]
        queue[i] = None
    front = -1
    rear -= 1

    # 빠져나가는 값 return
    return data 

def peek():
    global SIZE, queue, front, rear
    if isQueueEmpty():
        print('Queue is empty')
        return None
    return queue[front + 1]  # front+1 위치가 첫 번째 요소

# 실행 코드
if __name__ == '__main__':
    while True:
        select = input('삽입(I)/추출(E)/확인(V)/종료(Q) --> ').upper()

        if select == 'Q':
            break
        elif select == 'I':
            data = input(f'데이터 입력 > ')
            enQueue(data)
            print(f'큐 상태 --> {queue}')
        elif select == 'E':
            data = deQueue()
            print(f'추출 데이터 : {data}')
            print(f'큐 상태 : {queue}')
        elif select == 'V':
            data = peek()
            print(f'확인 데이터 : {data}')
            print(f'큐 상태 {queue}')
        else:
            print('입력 오류')


```

### 3. P309 편의점 판매된 물건 목록 출력하기 문제 역시 while True: 문으로 동적으로 입력받는 프로그램으로 변경하시오.
- 데이터의 종류는 [ 콘말차, 삿포로 맥주, 아카페라 벤티 헤이즐넛, 레어치즈푸딩, 척오리지널 메가 사워, 요아정 요거트컵, 페퍼로니피자주먹밥, 널담 슈톨렌, 딸기마시멜로케이크, 버니공쥬 딸기뚱카롱, 고추잡채호빵, 체다 슈크림붕어스낵 ]

```python
import random

class TreeNode:
    def __init__(self):
        self.left = None
        self.data = None
        self.right = None

memory = []
root = None
dataAry = []


# 데이터 입력
while True:
    data = input('데이터를 입력해주세요 > (종료시 Q 입력) ')
    if data == 'Q':
        break
    dataAry.append(data)

# 팔린 물건을 랜덤으로 
sellAry = [random.choice(dataAry) for _ in range(20)]
print(f'오늘 판매된 물건(중복 O) -> {sellAry}')


if __name__ == '__main__':
    node = TreeNode()
    node.data = sellAry[0]
    root = node
    memory.append(node) # 첫번째로 팔린 물건을 root node로 만들어서 memory에 추가

    
    for name in sellAry[1:]:
        node = TreeNode()
        node.data = name
        
        current = root
        while True:
            if name == current.data:  # 중복제거를 위해 값이 같다면 break
                break
            if name < current.data:  # 현재 값보다 지금 들어오는 name의 값이 작다면 왼쪽으로
                if current.left is None:    # 왼쪽이 비어있으면 그곳에 추가
                    current.left = node
                    memory.append(node)
                    break
                current = current.left # 값이 있다면 현재를 왼쪽으로 바꾼 다음 다시 while문
            else:  # 현재 값보다 지금 들어오는 name의 값이 크다면 오른쪽으로
                if current.right is None: # 오른쪽이 비어있으면 그곳에 추가
                    current.right = node
                    memory.append(node)
                    break
                current = current.right # 오른쪽에 값이 있다면 현재를 오른쪽으로 바꾼 다음 while문

    print('이진 탐색 트리 구성 완료!')

def preorder(node):
    if node is None:
        return
    print(node.data, end=' -> ')
    preorder(node.left)
    preorder(node.right)

print(f'오늘 판매된 물건(중복 X) -->', end=' ')
preorder(root)
```

### 4. P473 Self Study 13-2 의 문제. 이진 검색 알고리즘을 풀면서 검색 횟수도 같이 출력하도록 구현하세요.

```python
import random

# 이진 검색 알고리즘 
def binSearch(ary, data):
    pos = -1
    start = 0 
    end = len(ary) -1 
    search_count = 0

    while (start <= end):  # 항상 끝값보다 시작값이 작거나 같아야함
        mid = (start + end) // 2 # 중간값 인덱스계산
        if data == ary[mid]: ## 내가 찾는 값이면
            search_count += 1
            return mid, search_count # return
        elif data > ary[mid]: # 내가 찾는 값이 중간값보다 크다면
            start = mid +1  # 시작을 뒤로 (중앙값뒤로)
        else:   # 내가 찾는 값이 중간값보다 작다면
            end = mid -1 # 끝을 앞으로 (중앙값앞으로)

    return pos, search_count


dataAry = sorted([random.randint(0, 100000) for _ in range(100000)])
print(f'배열 일부 -> [{dataAry[:5]} ... [{dataAry[-5:]}]')

findData = random.randint(1,100000)

pos,search_count = binSearch(dataAry,findData)
if  pos == -1 :
    print(f'{findData} 없음')
else:
    print(f'{findData}(은)는 / {pos} 위치에 존재')
print(f'## {search_count}회 검색함')
```
